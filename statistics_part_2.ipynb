##statistics part 2-theory questions##

**1.** **What is hypothesis testing in statistics?**  
Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data.

**2.** **What is the null hypothesis, and how does it differ from the alternative hypothesis?**  
The **null hypothesis (H₀)** assumes no effect or difference, while the **alternative hypothesis (H₁)** assumes there is an effect or difference. Hypothesis testing checks if there's enough evidence to reject H₀ in favor of H₁.

**3.** **What is the significance level in hypothesis testing, and why is it important?**  
The **significance level (α)** is the threshold for rejecting H₀, commonly set at 0.05. It represents the probability of making a Type I error (rejecting a true H₀).

**4.** **What does a P-value represent in hypothesis testing?**  
The **P-value** is the probability of observing the sample results, or more extreme, if H₀ is true.

**5.** **How do you interpret the P-value in hypothesis testing?**  
If **P-value ≤ α**, reject H₀ (evidence supports H₁). If **P-value > α**, do not reject H₀ (not enough evidence for H₁).

**6.** **What are Type 1 and Type 2 errors in hypothesis testing?**  
- **Type I error**: Rejecting a true H₀.  
- **Type II error**: Failing to reject a false H₀.

**7.** **What is the difference between a one-tailed and a two-tailed test in hypothesis testing?**  
- **One-tailed**: Tests for an effect in one direction only.  
- **Two-tailed**: Tests for effects in both directions.

**8.** **What is the Z-test, and when is it used in hypothesis testing?**  
A **Z-test** is used when the population variance is known and the sample size is large (n ≥ 30).

**9.** **How do you calculate the Z-score, and what does it represent in hypothesis testing?**  
**Z = (X̄ - μ) / (σ / √n)**  
It measures how many standard deviations the sample mean is from the population mean.

**10.** **What is the T-distribution, and when should it be used instead of the normal distribution?**  
The **T-distribution** is used when the sample size is small (n < 30) and population standard deviation is unknown.

**11.** **What is the difference between a Z-test and a T-test?**  
- **Z-test**: Known population variance, large samples.  
- **T-test**: Unknown population variance, small samples.

**12.** **What is the T-test, and how is it used in hypothesis testing?**  
A **T-test** checks if the means of groups differ significantly when population variance is unknown.

**13.** **What is the relationship between Z-test and T-test in hypothesis testing?**  
Both test means; **T-test** is used when variance is unknown or sample is small; **Z-test** is for large samples with known variance.

**14.** **What is a confidence interval, and how is it used to interpret statistical results?**  
A **confidence interval** gives a range of values within which the population parameter is likely to fall, with a certain confidence level (e.g., 95%).

**15.** **What is the margin of error, and how does it affect the confidence interval?**  
The **margin of error** is the range added/subtracted from the estimate; a larger margin widens the interval, indicating less precision.

**16.** **How is Bayes' Theorem used in statistics, and what is its significance?**  
**Bayes’ Theorem** updates the probability of a hypothesis based on new evidence. It’s key in Bayesian statistics and decision-making.

**17.** **What is the Chi-square distribution, and when is it used?**  
The **Chi-square distribution** is used for categorical data, often in tests of independence or goodness of fit.

**18.** **What is the Chi-square goodness of fit test, and how is it applied?**  
It tests whether observed categorical data matches expected frequencies under a specific hypothesis.

**19.** **What is the F-distribution, and when is it used in hypothesis testing?**  
The **F-distribution** is used to compare variances and in **ANOVA tests**. It's skewed and non-negative.

**20.** **What is an ANOVA test, and what are its assumptions?**  
**ANOVA (Analysis of Variance)** tests differences between group means. Assumes:
- Independence  
- Normality  
- Equal variances

**21.** **What are the different types of ANOVA tests?**  
- **One-way ANOVA**: One independent variable.  
- **Two-way ANOVA**: Two independent variables.  
- **Repeated measures ANOVA**: Same subjects across conditions.

**22.** **What is the F-test, and how does it relate to hypothesis testing?**  
The **F-test** compares two variances and is the basis for ANOVA. It checks if group variances differ significantly.

**1**.**Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results**

```python
from scipy.stats import norm
import numpy as np

sample = np.random.normal(loc=102, scale=15, size=50)
pop_mean = 100
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)

z = (sample_mean - pop_mean) / (sample_std / np.sqrt(len(sample)))
p_value = 2 * (1 - norm.cdf(abs(z)))

print(f"Z = {z:.2f}")
print(f"P-value = {p_value:.4f}")

if p_value < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

**3.Implement a one-sample Z-test using Python to compare the sample mean with the population mean**

```python
def one_sample_z_test(sample, pop_mean):
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)
    z = (sample_mean - pop_mean) / (sample_std / np.sqrt(len(sample)))
    p_value = 2 * (1 - norm.cdf(abs(z)))
    return z, p_value

data = np.random.normal(55, 10, 100)
z_score, p = one_sample_z_test(data, 50)
print(f"Z-score: {z_score:.2f}, P-value: {p:.4f}")
```

**4.Perform a two-tailed Z-test using Python and visualize the decision region on a plotC**

```python
import matplotlib.pyplot as plt

sample = np.random.normal(105, 10, 30)
pop_mean = 100
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)

z = (sample_mean - pop_mean) / (sample_std / np.sqrt(len(sample)))

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)
plt.plot(x, y)

alpha = 0.05
z_crit = norm.ppf(1 - alpha/2)

plt.axvline(z_crit, color='r', linestyle='--', label='Critical value')
plt.axvline(-z_crit, color='r', linestyle='--')
plt.axvline(z, color='g', linestyle='-', label=f'Z = {z:.2f}')
plt.fill_between(x, 0, y, where=(x > z_crit) | (x < -z_crit), color='red', alpha=0.3)
plt.title("Two-tailed Z-Test Decision Region")
plt.legend()
plt.grid(True)
plt.show()
```

**5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing**

```python
def visualize_errors():
    x = np.linspace(-4, 6, 1000)
    h0 = norm.pdf(x, 0, 1)
    h1 = norm.pdf(x, 2, 1)

    plt.plot(x, h0, label='H0: μ = 0')
    plt.plot(x, h1, label='H1: μ = 2')

    alpha_region = x > norm.ppf(0.95)
    beta_region = x < norm.ppf(0.95, loc=2, scale=1)

    plt.fill_between(x, 0, h0, where=alpha_region, color='red', alpha=0.3, label='Type I Error')
    plt.fill_between(x, 0, h1, where=beta_region, color='blue', alpha=0.3, label='Type II Error')

    plt.title("Type I and Type II Errors")
    plt.legend()
    plt.grid(True)
    plt.show()

visualize_errors()
```

**9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance**

```python
def confidence_interval(sample, confidence=0.95):
    mean = np.mean(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(len(sample))
    margin = norm.ppf(1 - (1 - confidence)/2) * std_err
    return mean - margin, mean + margin

sample = np.random.normal(50, 10, 100)
ci = confidence_interval(sample)
print(f"95% Confidence Interval: {ci}")
```

**10.write a Python program to calculate the margin of error for a given confidence level using sample data**


```python
def margin_of_error(sample, confidence=0.95):
    std_err = np.std(sample, ddof=1) / np.sqrt(len(sample))
    z_crit = norm.ppf(1 - (1 - confidence)/2)
    return z_crit * std_err

sample = np.random.normal(100, 20, 60)
moe = margin_of_error(sample)
print(f"Margin of Error: {moe:.2f}")
```

**11.Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process**

```python
def bayes_theorem(p_a, p_b_given_a, p_b_given_not_a):
    p_not_a = 1 - p_a
    p_b = p_b_given_a * p_a + p_b_given_not_a * p_not_a
    p_a_given_b = (p_b_given_a * p_a) / p_b
    return p_a_given_b

# Example: Disease Test
p_disease = 0.01
p_positive_given_disease = 0.99
p_positive_given_no_disease = 0.05

posterior = bayes_theorem(p_disease, p_positive_given_disease, p_positive_given_no_disease)
print(f"Probability of disease given positive test: {posterior:.4f}")
```

```python
data = np.random.normal(100, 15, 40)
sample_var = np.var(data, ddof=1)
pop_var = 225  # σ² = 15²

chi2_stat = (len(data) - 1) * sample_var / pop_var
p_val = 2 * min(chi2.cdf(chi2_stat, df=len(data)-1), 1 - chi2.cdf(chi2_stat, df=len(data)-1))

print(f"Chi-square statistic: {chi2_stat:.2f}, P-value: {p_val:.4f}")
```
